```python
def to_float(number_str):
    if number_str.endswith("k"):
        return float(number_str[:-1]) * 1000
    elif number_str.endswith("M"):
        return float(number_str[:-1]) * 1_000_000
    elif number_str.endswith("+"):
        return float(number_str[:-1].replace(",", ""))    
    elif number_str.startswith("$"):
        return float(number_str.replace("$", ""))    
    else:
        return float(number_str)
```

```python
# df["Reviews"] = df["Reviews"].apply(to_float).apply(int)
df.loc[:, "Reviews"] = df.loc[:, "Reviews"].apply(to_float).apply(int)
```

```python
df_excluding_no_price = df[df["Price"] != "Everyone"]
```

```python
df_excluding_no_price.loc[:, "Price"] = df_excluding_no_price.loc[:, "Price"].apply(to_float)
```

```python
df_excluding_no_size = df_excluding_no_price[df_excluding_no_price["Size"] != "Varies with device"]
```

```python
# df_excluding_no_size["Size"] = df_excluding_no_size["Size"].apply(to_float)
df_excluding_no_size.loc[:, "Size"] = df_excluding_no_size.loc[:, "Size"].apply(to_float)
```

```python
# df_excluding_no_size.groupby(["Category"]).describe()
df_excluding_no_size.groupby(["Category"])["Rating", "Price"].describe()
```
