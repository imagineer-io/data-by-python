# 데이터 가공하기

## 문자를 실수로 만들어보자

```python
def to_float(number_str):
    if number_str.endswith("k"): # "17k" = 17000
        return float(number_str[:-1]) * 1000 # h = "hello" # h[1:] = "ello" # h[:-1] = "hell"
    elif number_str.endswith("M"): # "17M" = 17000000
        return float(number_str[:-1]) * 1_000_000
    elif number_str.endswith("+"):
        return float(number_str[:-1].replace(",", "")) # "500,000".replace(",", "") == "500000"
    elif number_str.startswith("$"):
        return float(number_str.replace("$", ""))    
    else:
        return float(number_str)
```

## apply는 무엇인가?

```python
df.loc[:, "Reviews"].apply(to_float)[:20]
```

## Dataframe 업데이트 하기

```python
# df["Reviews"] = df["Reviews"].apply(to_float).apply(int)
df.loc[:, "Reviews"] = df.loc[:, "Reviews"].apply(to_float).apply(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.loc[:, "Price"] = df.loc[:, "Price"].apply(to_float)
```

```python
# df_excluding_no_size["Size"] = df_excluding_no_size["Size"].apply(to_float)
df.loc[:, "Size"] = df.loc[:, "Size"].apply(to_float)
```

```python
df.groupby(["Category"]).mean()
```

```python
# df.groupby(["Category"]).describe()
df.groupby(["Category"])["Rating", "Price"].describe()
```

# 그래프의 기본 요소

<img src="sphx_glr_anatomy_001.png" alt="anatomy" style="width: 500px;"/>

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## markers
https://matplotlib.org/api/markers_api.html

```python
plt.plot(x, y, marker=".")
```

## linestyle
https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html

```python
plt.plot(x, y, ls='--')
```

## text & annotation

```python
plt.text(2, 0, 'cosine graph', style='italic')
plt.annotate("cosine",
            xy=(8, 0),
            xycoords='data',
            xytext=(12, 0), textcoords='data', arrowprops=dict(arrowstyle="->",
            connectionstyle="arc3"),)
plt.plot(x, y)
```

## legend
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html

```python
plt.plot(x, y, label="cosine")
plt.legend(loc="best")
```

## label

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```
